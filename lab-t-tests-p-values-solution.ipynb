```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1.One tailed t-test 
```

```python
machine = pd.read_csv('machine.txt',encoding = "utf-16", sep = "\t", names = ["New", "Old"], header=0)
```

```python
machine
```

```python
#H0: new > old
#h1: new < old
new_mean=machine['New'].mean()
old_mean=machine['Old'].mean()
print('p value (one tailed): ',st.ttest_1samp(machine['New'], old_mean)[1]/2)
```

```python
#p value < 0.025,so we reject H0 here.New mathine is faster than old machine.
```

```python
#2.Matched Pairs Test
```

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon
```

```python
#H0 Attach = Defence 
#H1 Attach != Defence
st.ttest_rel(pokemon['Attack'], pokemon['Defense'])
```

```python
#Since P value < 0.05,we reject H0,so the defense and attack scores are not equal.
```

```python
#3.Inferential statistics - ANOVA
```

```python
beam = pd.read_excel('anova_lab_data.xlsx',names = ["Power", "Etching Rate"],header=0)
```

```python
power = beam['Power'].apply(lambda x: int(x.split(" ")[0]))#here we remove W in the power column
```

```python
beam['Power']=power
```

```python
beam
```

```python
beam['power_count'] = beam.groupby('Power').cumcount()
```

```python
beam_pivot = beam.pivot(index='power_count',columns='Power',values='Etching Rate')
```

```python
beam_pivot.columns = ['Power_'+str(x) for x in beam_pivot.columns.values]
```

```python
beam_pivot
```

```python
#1)State the null hypothesis
#H0:the different power leads to same etching rate
#2)State the alternate hypothesis
#H1: different power leads to different etching rate
```

```python
st.f_oneway(beam_pivot.Power_160,beam_pivot.Power_180,beam_pivot.Power_200)
```

```python
#What is the significance level?-->here also we will take it to be 0.05
#What are the degrees of freedom of model, error terms, and total DoF
```

```python
The pvalue < 0.05,we reject H0.So the different power leads to different etching rate.
```
